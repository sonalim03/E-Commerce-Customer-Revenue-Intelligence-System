```python
#  CLEAN STRINGS (remove quotes, standardize case)
for col in ['customer_id', 'customer_unique_id', 'customer_city', 'customer_state']:
    customers[col] = customers[col].astype(str).str.strip('"').str.strip()
```

```python
# CLEAN ZIP CODES
customers['customer_zip_code_prefix'] = (
    customers['customer_zip_code_prefix']
    .astype(str)
    .str.replace(r'\D', '', regex=True)
    .str.zfill(5)
)
```

```python
print("\n" + "="*50)
print("DATA QUALITY CHECKS")
print("="*50)

# 1️⃣ Invalid States (must be exactly 2 uppercase letters)
print("\nInvalid states (should be 2 letters):")
invalid_states = customers[
    ~customers['customer_state'].str.match(r'^[A-Z]{2}$', na=False)
]
print(invalid_states[['customer_state']].value_counts() if len(invalid_states) > 0 else "None")


# 2️⃣ Cities with less than 3 characters
print("\nCities with < 3 characters:")
invalid_cities = customers[
    customers['customer_city'].str.len() < 3
]
print(invalid_cities[['customer_city']].value_counts() if len(invalid_cities) > 0 else "None")


# 3️⃣ Invalid ZIP codes (must be exactly 5 digits)
print("\nInvalid ZIP codes (should be 5 digits):")
invalid_zips = customers[
    ~customers['customer_zip_code_prefix'].str.match(r'^\d{5}$', na=False)
]
print("Invalid ZIP count:", len(invalid_zips))
```

```python
#  HANDLE DUPLICATES
duplicate_ids = customers.duplicated('customer_id').sum()
duplicate_unique = customers.duplicated('customer_unique_id').sum()

print(f"\nDuplicate customer_ids: {duplicate_ids}")
print(f"Customers with multiple orders: {duplicate_unique}")
```

```python
# 6. ADD USEFUL FEATURES
# Add region based on state
region_map = {
    'AC':'Norte','AP':'Norte','AM':'Norte','PA':'Norte','RO':'Norte','RR':'Norte','TO':'Norte',
    'AL':'Nordeste','BA':'Nordeste','CE':'Nordeste','MA':'Nordeste','PB':'Nordeste','PE':'Nordeste',
    'PI':'Nordeste','RN':'Nordeste','SE':'Nordeste','DF':'Centro-Oeste','GO':'Centro-Oeste',
    'MT':'Centro-Oeste','MS':'Centro-Oeste','ES':'Sudeste','MG':'Sudeste','RJ':'Sudeste',
    'SP':'Sudeste','PR':'Sul','RS':'Sul','SC':'Sul'
}
customers['region'] = customers['customer_state'].map(region_map)
```

```python
import pandas as pd
import numpy as np

orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
orders = orders[orders['order_status'] == 'delivered']
snapshot = orders['order_purchase_timestamp'].max() + pd.Timedelta(days=1)

rfm = (
    orders
    .merge(customers, on='customer_id')
    .merge(order_payments[['order_id', 'payment_value']], on='order_id')
    .groupby('customer_unique_id')
    .agg(
        Recency   = ('order_purchase_timestamp', lambda x: (snapshot - x.max()).days),
        Frequency = ('order_id', 'nunique'),
        Monetary  = ('payment_value', 'sum')
    )
    .reset_index()
)

repeat_mask = rfm['Frequency'] > 1
rfm.loc[repeat_mask, 'R'] = pd.qcut(rfm.loc[repeat_mask, 'Recency'], q=4, labels=[4,3,2,1]).astype(float)
rfm.loc[repeat_mask, 'F'] = pd.qcut(rfm.loc[repeat_mask, 'Frequency'].rank(method='first'), q=4, labels=[1,2,3,4]).astype(float)
rfm.loc[repeat_mask, 'M'] = pd.qcut(rfm.loc[repeat_mask, 'Monetary'].rank(method='first'), q=4, labels=[1,2,3,4]).astype(float)

def rfm_segment(row):
    if row['Frequency'] == 1:   return 'Single Purchase'
    r, f = int(row['R']), int(row['F'])
    if r >= 3 and f >= 3:       return 'Champions'
    if r <= 2 and f >= 3:       return 'At Risk Loyal'
    if r >= 3 and f <= 2:       return 'Recent Low Frequency'
    return 'Regular Repeat Buyers'

rfm['Segment'] = rfm.apply(rfm_segment, axis=1)

rfm_summary = (
    rfm.groupby('Segment')
    .agg(customers=('customer_unique_id','count'), avg_ltv=('Monetary','mean'),
         avg_orders=('Frequency','mean'), avg_recency=('Recency','mean'))
    .round(2).sort_values('avg_ltv', ascending=False)
)
print(rfm_summary)
```

```python
# Seller Rating Effect on Customer Retention
# Seller Rating Effect on Customer Retention
import pandas as pd

seller_avg_rating = (
    reviews
    .merge(order_items[['order_id','seller_id']], on='order_id')
    .groupby('seller_id')['review_score']
    .mean()
    .reset_index()
    .rename(columns={'review_score': 'avg_seller_rating'})
)

cust_total = (
    orders.merge(customers, on='customer_id')
    .groupby('customer_unique_id')
    .size().reset_index(name='total')
)

seller_merged = (
    orders
    .merge(order_items[['order_id','seller_id']], on='order_id')
    .merge(customers, on='customer_id')
    .merge(seller_avg_rating, on='seller_id')
    .merge(cust_total, on='customer_unique_id')
)

seller_merged['repeat'] = seller_merged['total'] > 1
seller_merged['rating_bin'] = pd.cut(
    seller_merged['avg_seller_rating'],
    bins=[0, 2, 3, 4, 5],
    labels=['1-2 (poor)', '2-3 (below avg)', '3-4 (good)', '4-5 (excellent)']
)

seller_result = (
    seller_merged.groupby('rating_bin', observed=False)['repeat']
    .agg(repeat_rate='mean', customers='count')
    .assign(repeat_rate=lambda d: (d['repeat_rate']*100).round(2))
)
print(seller_result)
```

```python
# Price Sensitivity (AOV vs. Retention)
import pandas as pd

aov = (
    order_items
    .merge(orders, on='order_id')
    .merge(customers, on='customer_id')
    .groupby('customer_unique_id')
    .agg(
        avg_order_value=('price','mean'),
        total_orders=('order_id','nunique')
    )
    .reset_index()
)

aov['segment'] = pd.cut(
    aov['avg_order_value'],
    bins=[0,100,300,1e6],
    labels=['Low','Medium','High']
)

result = (
    aov.groupby('segment', observed=False)
    .agg(
        repeat_rate=('total_orders', lambda x: (x>1).mean()),
        customers=('customer_unique_id','count')
    )
)

print(result)
```
